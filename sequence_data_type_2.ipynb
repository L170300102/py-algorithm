```python
def erasing(string):
    s1 = string.split()
    s2 = "".join(s1)
    return s2

s1 = "hello python notebook and jupyter"
res1 = erasing(s1)
print(res1)
```

## 알파벳순 모든 단어 중복제거 횟수 반환

```python
import string
import sys

def count_unique_word(filename):
    words={}
    strip = string.whitespace + string.punctuation + string.digits + "\"'"
#     for filename in sys.argv[1:]:
    with open(filename) as file:
        for line in file:
            for word in line.lower().split():
                word = word.strip(strip)
                if len(word) > 2:
                    words[word] = words.get(word, 0) + 1
    for word in sorted(words):
        print(f"{word}: {words[word]} times")
```

```python
count_unique_word('alice_in_wonderland.txt')
```
